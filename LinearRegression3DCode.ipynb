# Testing 2 input and 1 output Linear Regression

```python
import numpy as np
import pandas as pd
data=pd.read_csv("data.csv")
data.columns=["input1","input2"]
data["output"]=(2*data["input1"]+3*data["input2"])/5 #randomly adding the third column and woh output treat
print(data)
```

Now convert this to numpy 2D array as processing is easier waha

```python
data=np.array(data)
print(data)
```

```python
input=data[:,0:2] #input mei all rows/datapoints ke 1st 2 columns
output=data[:,2] #output mei last column
print(input)
print(output)
```

Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,output,random_state=1)
```

Now, inbuilt Linear Regression use karenge

```python
from sklearn.linear_model import LinearRegression
algorithm=LinearRegression()
algorithm.fit(x_train,y_train)
print("m1 : ",algorithm.coef_[0])
print("m2 : ",algorithm.coef_[1])
print("m3 : ",algorithm.intercept_)
```

Clearly our output was 2/5*x1+3/5*x2+0 joh woh predict baas itne datas se joh really impressive
issi ko kehte output has linear relation wrt x1 and x2 so linear regression se solved

```python
print("Score : ",algorithm.score(x_test,y_test))
```

So score 1 aaya as exact match kar rhe testing output se y_predicted joh ki sahi as m1 and m2 exact hi nikal liya

```python
def get_optimal_coefficient(x_train,y_train):
    '''
    Returns coef_ array [m1,m2,m3]
    Refer copy for mapping
    '''
    x1=x_train[:,0]
    x2=x_train[:,1]
    y=y_train
    a=(x1*x2).mean()
    b=(x1**2).mean()
    c=(x2**2).mean()
    d=x1.mean()
    e=x2.mean()
    f=(x1*y).mean()
    g=(x2*y).mean()
    h=y.mean()
    m2=((d*h-f)*(d*e-a)-(d**2-b)*(e*h-g))/((c-e**2)*(d**2-b)+(d*e-a)**2)
    m1=(e*h-g-(e**2-c)*m2)/(e*d-a)
    m3=h-d*m1-e*m2
    return (m1,m2,m3)
```

```python
m1,m2,m3=get_optimal_coefficient(x_train,y_train)
print("m1 : ",m1)
print("m2 : ",m2)
print("m3 : ",m3)
```

Shows that our calculation is perfect!!!!!!!!!!!!!

```python
def coefficient_of_determination(y_predicted,y_actual):
    u=((y_actual-y_predicted)**2).sum()
    v=((y_actual-y_actual.mean())**2).sum()
    return 1-u/v
```

```python
y_predicted=m1*x_test[:,0]+m2*x_test[:,1]+m3
print("Score : ",coefficient_of_determination(y_predicted,y_test))
```
